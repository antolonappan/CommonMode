```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_pickle('tod.pkl')
```

```python
bolo_time = df['bolotime']*3600.*24
```

```python
df_tod = df.loc[:, df.columns!='bolotime']
signal = df_tod.values.T
```

```python
def sample_rate(time):
    return 1/np.mean(np.diff(time))
```

```python
sample_rate(bolo_time)
```

```python
cov = df_tod.cov().values
```

```python
E_,R_ = np.linalg.eig(cov)
E_[np.isnan(E)] = 0.
idx = np.argsort(-E_)
E = E_[idx]
R = R_[:, idx]
```

```python
E_mat = np.zeros([len(E),len(E)])
np.fill_diagonal(E_mat, E)
```

```python
#C = np.dot(np.dot(V, E_mat),V.T)
```

```python
plt.plot(E)
plt.semilogy()
```

```python
nmodes = 4
R_n = R[:, :nmodes]
modes = np.dot(R_n.transpose(), signal)
```

```python
signal_cm = np.zeros_like(signal)
```

```python
for i in range(len(E)):
    signal_cm[i] = signal[i] - np.dot(R_n[i, :], modes)
```

```python
cov_cm = np.cov(signal_cm)
```

```python
Ecm, Vcm = np.linalg.eig(cov_cm)
Ecm[np.isnan(Ecm)] = 0.
idx = np.argsort(-Ecm)
Ecm = Ecm[idx]
```

```python
plt.plot(E[:100])
plt.plot(Ecm[:100])
plt.semilogy()
```

```python
det = list(df_tod.columns)
```

```python
df_cm = pd.DataFrame(signal_cm.T)
df_cm.columns = det    
```

```python
sns.heatmap(df_cm.corr(),vmin=-1, vmax=1, center=0,
            cmap=sns.diverging_palette(20, 220, n=200),
            cbar_kws ={'shrink': .6, 'label': 'Correlation'},
            square=True)
```
